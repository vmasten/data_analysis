# Video Game Statistics

```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
df = pd.read_csv('./vgsales.csv')
```

# Which company is the most common publisher?

```python
high_sales = (df.Global_Sales).groupby(df.Publisher)
series = high_sales.count()
series.sort_values(ascending=False).head(10)
```

# What's the most common platform?

```python
platform_count = (df.Global_Sales).groupby(df.Platform)
series = platform_count.count()
series.sort_values(ascending=False).head(1)
```

# What's the most popular genre?

```python
popular_genre = df.Global_Sales.groupby(df.Genre)
series = popular_genre.count()
series.sort_values(ascending=False)
```
